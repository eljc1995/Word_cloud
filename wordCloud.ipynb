```python
import wikipedia
```

```python
from wordcloud import WordCloud
```

```python
import matplotlib.pyplot as plt 
```

```python
info = wikipedia.summary('data science ')
info
```

```python
wordCloud=WordCloud().generate(info)
```

```python
plt.imshow(wordCloud,interpolation = 'bilinear')
plt.axis('off')
plt.show
```

```python
image = wordCloud.to_image()
image.save('juancarlosospinaarangodatascientist.png')
image.show()
```
